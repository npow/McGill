```python
import matplotlib.cm as cm
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline
import math
import numpy as np
import os
import sys
from PIL import Image

DATA_DIR = 'data/data'
```

```python
def mse(A, B):
    return ((A-B)**2).mean()

def euclidean(A, B):
    return np.linalg.norm(A-B)

class PCA(object):
    def __init__(self, n_components):
        self.k = n_components
        self.explained_variance_ratio_ = None
        self.components_ = None
        self.mean_ = None
        
    def fit(self, X):
        U, S, V = np.linalg.svd(X, full_matrices=False)
        self.explained_variance_ratio_ = (S**2) / X.shape[0]
        self.explained_variance_ratio_ /= self.explained_variance_ratio_.sum()
        self.components_ = V[:self.k]  
        self.explained_variance_ratio_ = self.explained_variance_ratio_[:self.k]

    def fit_transform(self, X):
        self.fit(X)
        return self.transform(X)
    
    def transform(self, X):
        return X.dot(self.components_.T)
    
    def inverse_transform(self, X):
        return X.dot(self.components_)
```

```python
def get_mean(fnames):
    width, height = Image.open(fnames[0]).size
    mean = np.zeros((width, height))
    for fname in fnames:
        mean += np.array(Image.open(fname), dtype=np.float) / len(fnames)
    return np.array(np.round(mean), dtype=np.uint8)

def plot_pca_components(X):
    L = []
    ks = [5] + range(10, 128, 10)
    for k in ks:
        print "*** k: %d" % k      
        sys.stdout.flush()
        select = PCA(n_components=k)
        select.fit(X)
        L.append(select.explained_variance_ratio_)
        if k == 10:
            for i in xrange(k):
                print i, select.components_[i]
    plt.figure()
    plt.xlabel('n_components')
    plt.ylabel('explained variance')
    plt.plot(ks, [x.sum() for x in L])
    plt.show()
    
def plot_pca_reconstruction(X_train, X_test):
    n = 3 # use first n images
    X_train = X_train[range(n),:] 
    X_test = X_test[range(n),:]
    train_L, test_L = [], []
    ks = [1, 2, 5, 10, 20]
    for k in ks:
        print "*** k: %d" % k 
        sys.stdout.flush()
        select = PCA(n_components=k)
        X_train_select = select.fit_transform(X_train)
        X_train_proj = select.inverse_transform(X_train_select)
        
        X_test_select = select.transform(X_test)
        X_test_proj = select.inverse_transform(X_test_select)
        
        train_errs, test_errs = [], []
        plt.figure()
        print "train images:"
        for i in xrange(n):
            plt.subplot(1, n, i)
            plt.imshow(X_train_proj[i].reshape(-1, 128))            
            train_errs.append(math.sqrt(mse(X_train[i], X_train_proj[i])))
        plt.show()

        plt.figure()
        print "test images:"
        for i in xrange(n):
            plt.subplot(1, n, i)
            plt.imshow(X_test_proj[i].reshape(-1, 128))            
            test_errs.append(math.sqrt(mse(X_test[i], X_test_proj[i])))  
        plt.show()
            
        train_avg_rmse = np.array(train_errs).mean()
        test_avg_rmse = np.array(test_errs).mean()
        
        train_L.append(train_avg_rmse)
        test_L.append(test_avg_rmse)
        print "k: ", k
        print "train_avg_rmse: ", train_avg_rmse
        print "test_avg_rmse: ", test_avg_rmse
        print "\n\n\n"

    plt.figure()
    plt.title('train images, RMSE vs n_components')
    plt.xlabel('n_components')
    plt.ylabel('RMSE')
    #plt.yscale('log')
    plt.plot(ks, train_L)
    plt.show()
    print "ks: ", ks
    print "train_L: ", train_L
    
    plt.figure()
    plt.title('test images, RMSE vs n_components')
    plt.xlabel('n_components')
    plt.ylabel('RMSE')
    #plt.yscale('log')
    plt.plot(ks, test_L)
    plt.show()
    print "ks: ", ks
    print "test_L: ", test_L    
    
def find_closest(X, z):
    closest = float('inf')
    retval = None
    for i,x in enumerate(X):
        dist = euclidean(x, z)
        if dist < closest:
            closest = dist
            retval = i
    return retval, closest
    
def do_classification(X_train, X_test, Y_train, Y_test, train_fnames, test_fnames):
    select_train = PCA(n_components=10)
    X_train_select = select_train.fit_transform(X_train)
    
    select_test = PCA(n_components=10)
    X_test_select = select_test.fit_transform(X_test)
    
    for i,x in enumerate(X_test_select):
        result, dist = find_closest(X_train_select, x)
        
        plt.figure() 
        plt.subplot(1, 4, 1)
        plt.axis('off')        
        plt.title('test img:\n %s' % test_fnames[i])
        plt.imshow(X_test[i].reshape(128, -1))
        
        plt.subplot(1, 4, 2)
        plt.axis('off')        
        plt.title('closest match:\n %s' % train_fnames[result])
        plt.imshow(X_train[result].reshape(128, -1))
        
        plt.subplot(1, 4, 3)
        plt.axis('off')        
        plt.title('raw test:\n ' + test_fnames[i])
        plt.imshow(mpimg.imread("%s/test/%s" % (DATA_DIR, test_fnames[i])), cmap=cm.Greys_r)
        
        plt.subplot(1, 4, 4)
        plt.axis('off')        
        plt.title('raw train:\n ' + train_fnames[result])
        plt.imshow(mpimg.imread("%s/training/%s" % (DATA_DIR, train_fnames[result])), cmap=cm.Greys_r)        
        
        plt.show()
        
        print "test img: ", test_fnames[i]
        print "closest: ", train_fnames[result]
        print "dist: %f" % dist
        print "\n\n\n"
        
def load_data(dataset):
    train_fnames = ["%s/training/%s" % (DATA_DIR, x) for x in filter(lambda x: x.find(dataset) > -1, os.listdir("%s/training" % DATA_DIR))]
    test_fnames = ["%s/test/%s" % (DATA_DIR, x) for x in filter(lambda x: x.find(dataset) > -1, os.listdir("%s/test" % DATA_DIR))]
    mean = get_mean(train_fnames) 
    plt.imshow(mean)
    plt.show()
        
    X_train = []
    for fname in train_fnames:
        X_train.append((Image.open(fname)-mean).reshape((1,-1)))
    X_train = np.vstack(X_train)
    X_test = []
    for fname in test_fnames:
        X_test.append((Image.open(fname)-mean).reshape((1,-1)))
    X_test = np.vstack(X_test)
    return X_train.astype(np.float), X_test.astype(np.float), train_fnames, test_fnames

def process(X_train, X_test):
    plot_pca_components(X_train)
    plot_pca_reconstruction(X_train, X_test)
```

```python
horse_X_train, horse_X_test, horse_train_fnames, horse_test_fnames = load_data('horse')
print horse_X_train.shape
print horse_X_test.shape
process(horse_X_train, horse_X_test)
```

```python
face_X_train, face_X_test, face_train_fnames, face_test_fnames = load_data('face')
process(face_X_train, face_X_test)
```

```python
X_train = np.concatenate([face_X_train, horse_X_train])
Y_train = np.zeros((X_train.shape[0]))
Y_train[X_train.shape[0]/2:] = 1
train_fnames = [x.split('/')[-1] for x in face_train_fnames + horse_train_fnames]

X_test = np.concatenate([face_X_test, horse_X_test])
Y_test = np.zeros((X_test.shape[0]))
Y_test[X_test.shape[0]/2:] = 1
test_fnames = [x.split('/')[-1] for x in face_test_fnames + horse_test_fnames]
    
do_classification(X_train, X_test, Y_train, Y_test, train_fnames, test_fnames)
```
