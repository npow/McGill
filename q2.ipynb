```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import os
import pyprind
from PIL import Image
from scipy.spatial.distance import euclidean
from sklearn.decomposition import PCA
from sklearn.externals import joblib
from sklearn.metrics import mean_squared_error as mse

DATA_DIR = 'data/data'
DATASETS = ['face', 'horse']
USE_CACHE = False
```

```python
def get_mean(fnames):
    width, height = Image.open(fnames[0]).size
    mean = np.zeros((width, height))
    for fname in fnames:
        mean += np.array(Image.open(fname), dtype=np.float) / len(fnames)
    return np.array(np.round(mean), dtype=np.uint8)

def plot_pca_components(dataset, X):
    L = []
    ks = range(10, 128)
    bar = pyprind.ProgBar(len(ks), monitor=True)
    for k in ks:
        if USE_CACHE:
            select = joblib.load('blobs/%s/select_%d.pkl' % (dataset, k))
        else:
            select = PCA(n_components=k)
            X_select = select.fit_transform(X)
            joblib.dump(select, 'blobs/%s/select_%d.pkl' % (dataset, k))
        L.append(select.explained_variance_ratio_)
        bar.update()
    plt.figure()
    plt.xlabel('n_components')
    plt.ylabel('explained variance')
    plt.plot(ks, [x.sum() for x in L])
    plt.show()
    
def plot_pca_reconstruction(dataset, X, isTest):
    X = X[range(3),:] # use first n images
    L = []
    ks = [1, 2, 5, 10, 20]
    for k in ks:
        print "*** k: %d" % k        
        if USE_CACHE or isTest:          
            select = joblib.load('blobs/%s/select_%d.pkl' % (dataset, k))
            if isTest:
                X_select = select.transform(X)
                X_proj = select.inverse_transform(X_select)                
            else:
                X_select = joblib.load('blobs/%s/X_select_%d.pkl' % (dataset, k))
                X_proj = joblib.load('blobs/%s/X_proj_%d.pkl' % (dataset, k))  
        else:
            select = PCA(n_components=k)
            X_select = select.fit_transform(X)
            X_proj = select.inverse_transform(X_select)
            joblib.dump(X_select, 'blobs/%s/X_select_%d.pkl' % (dataset, k))
            joblib.dump(X_proj, 'blobs/%s/X_proj_%d.pkl' % (dataset, k))
            joblib.dump(select, 'blobs/%s/select_%d.pkl' % (dataset, k))
        errs = []
        plt.figure()
        for i in xrange(X.shape[0]):
            #plt.subplot(1, X.shape[0], 1)
            #plt.imshow(X[i].reshape(-1, 128))
            plt.subplot(1, X.shape[0], i)
            plt.imshow(X_proj[i].reshape(-1, 128))            
            errs.append(math.sqrt(mse(X[i], X_proj[i])))
        plt.show()
        avg_rmse = np.array(errs).mean()
        L.append(avg_rmse)
        print "k: ", k, " avg_rmse: ", avg_rmse, "\n\n\n"

    plt.figure()
    plt.title('RMSE vs n_components')
    plt.xlabel('n_components')
    plt.ylabel('RMSE')
    #plt.yscale('log')
    plt.plot(ks, L)
    plt.show()
    print ks, L
    
def find_closest(X, z):
    closest = float('inf')
    retval = None
    for i,x in enumerate(X):
        dist = euclidean(x, z)
        if dist < closest:
            closest = dist
            retval = i
    return retval, closest
    
def do_classification():
    train_fnames = os.listdir("%s/training" % DATA_DIR)
    test_fnames = os.listdir("%s/test" % DATA_DIR)  
    
    X_train = np.concatenate([joblib.load('blobs/%s/X_train.pkl' % s) for s in ['face', 'horse']], axis=0)
    Y_train = np.zeros((X_train.shape[0]))
    Y_train[X_train.shape[0]/2:] = 1
    
    X_test = np.concatenate([joblib.load('blobs/%s/X_test.pkl' % s) for s in ['face', 'horse']], axis=0)
    Y_test = np.zeros((X_test.shape[0]))
    Y_test[X_test.shape[0]/2:] = 1    
    
    select_train = PCA(n_components=10)
    X_train_select = select_train.fit_transform(X_train)
    
    select_test = PCA(n_components=10)
    X_test_select = select_test.fit_transform(X_test)
    
    for i,x in enumerate(X_test_select):
        result, dist = find_closest(X_train_select, x)
        plt.figure()
        plt.suptitle('dist: %f' % dist)
        plt.subplot(1, 2, 1)
        plt.title('test img: %s' % test_fnames[i])
        plt.imshow(X_test[i].reshape(128, -1))
        plt.subplot(1, 2, 2)
        plt.title('closest match: %s' % train_fnames[result])
        plt.imshow(X_train[result].reshape(128, -1))
        plt.show()

def process(dataset):
    if USE_CACHE:
        X_train = joblib.load('blobs/%s/X_train.pkl' % dataset)
        X_test = joblib.load('blobs/%s/X_test.pkl' % dataset)
        mean = joblib.load('blobs/%s/mean.pkl' % dataset)        
    else:
        train_fnames = ["%s/training/%s" % (DATA_DIR, x) for x in filter(lambda x: x.find(dataset) > -1, os.listdir("%s/training" % DATA_DIR))]
        test_fnames = ["%s/test/%s" % (DATA_DIR, x) for x in filter(lambda x: x.find(dataset) > -1, os.listdir("%s/test" % DATA_DIR))]    
        mean = get_mean(train_fnames)
        X_train = np.array([(Image.open(fname)-mean).ravel() for fname in train_fnames])
        X_test = np.array([(Image.open(fname)-mean).ravel() for fname in test_fnames])
        joblib.dump(X_train, 'blobs/%s/X_train.pkl' % dataset)
        joblib.dump(X_test, 'blobs/%s/X_test.pkl' % dataset)
        joblib.dump(mean, 'blobs/%s/mean.pkl' % dataset)
    
    plt.imshow(mean)
    
    plot_pca_components(dataset, X_train)
    
    print "***** TRAIN *****"
    plot_pca_reconstruction(dataset, X_train, isTest=False)
    
    print "***** TEST *****"
    plot_pca_reconstruction(dataset, X_test, isTest=True)
```
